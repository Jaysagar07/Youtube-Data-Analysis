Analysis 1: How many no. of videos are from different years?

```python
#Creating separate column for trending year
data['trending_year'] = data['trending_date'].apply(lambda x: '20' + x[0:2])
```

Analysis 2: what are the top tags of trending videos


```python
data.head()
```

```python
#Installing wordcloud
pip install wordcloud
```

```python
#Word Cloud is a data visualization technique used for representing text data in which the 
#size of each word indicates its frequency

from wordcloud import WordCloud
```

```python
#joining the tags of all videos
tags=' '.join(data['tags'])
```

```python
import re
```

```python
#removing un-necessary data from tags
tag=re.sub('[^a-zA-Z]',' ',tags)
tag=re.sub(' +',' ',tag)
```

```python
#Initiating WordCloud plot
wordcloud=WordCloud(width=1000,height=1000).generate(tag)
```

```python
#Creating Wordcloud
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.show()
```

Analysis 3: How many views do our trending videos have?

```python
data.head()
```

```python
import seaborn as sns
```

```python
#Visualizing using distribution plot
sns.distplot(data[data['views']<25e6]['views'],kde=False)
plt.xlabel('views')
plt.ylabel('number of videos')
plt.show()
```

```python
#calculating % of videos having less than 5 million views
data[data['views']<5e6]['views'].count()/data['views'].count()*100
```

Analysis 4: which channel have the largest number of trending videos?

```python
#Creating separate dataframe
video_count = data.groupby('channel_title').size().reset_index(name='video_count')
```

```python
video_count.head()
```

```python
#Sorting on video count
video_count.sort_values(by='video_count',ascending=False).head(10)
```

```python
#visualizing using barplot
plt.figure()
sns.barplot(x='channel_title',y='video_count',data=video_count)
plt.xticks(rotation=90)
plt.show()
```

Analysis 6: When were trending videos published? at which times of the day?

```python
data.head()
```

```python
#Creating a new column for publish hour of video
data['publish_hour'] = data['publish_time'].str[11:13]
```

```python
#Creating separate dataframe
publish_time=data['publish_hour'].value_counts().to_frame().reset_index()
```

```python
publish_time
```

```python
publish_time=publish_time.rename(columns={'index':'Hours','publish_hour':'Counts'})
```

```python
#Visualizing using barplot
plt.figure()
sns.barplot(x='Hours',y='Counts',data=publish_time)
plt.show()
```
